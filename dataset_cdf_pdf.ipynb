```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris['species'].value_counts()
```

```python
vr = iris[iris['species']=="virginica"]
vs = iris[iris['species']=='versicolor']
st = iris[iris['species']=='setosa']
```

#  PDF (Virginica)

    Bins = 10  interval sayıdır.
    counts = array([ 1,  5, 12,  4,  9,  8,  5,  2,  1,  3], dtype=int64)
    bin_edges = array([4.5 , 4.74, 4.98, 5.22, 5.46, 5.7 , 5.94, 6.18, 6.42, 6.66, 6.9 ]) 
    bin_edges dəyişənində həmişə bins+1 sayda element olur, çünki 10 intervalın yaranması üçün 11 element lazımdır.
    counts bin_edges- dəki hər intervala düşən element sayıdır: 4.5-4.74 aralığında 1, 4.98-5.22 aralığında isə 12 iris var 
 

```python
counts, bin_edges = np.histogram(vr['petal_length'], bins = 10)
pdf = counts/(sum(counts))
plt.legend(('PDF'))
plt.title('PDF and PDF For iris virginica')
plt.xlabel("Petal length")
plt.ylabel("Percentage")
#bin_edges-dəki element sayı pdf-dən 1 vahid çox olduğundan bin_edgesdəki ilk elementi götürmürük
plt.plot(bin_edges[1:],pdf)
```

# CDF (Virginica)

```python
counts, bin_edges = np.histogram(vr['petal_length'], bins = 10)

pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)


plt.title('CDF For iris virginica')
plt.xlabel("Petal length")
plt.ylabel("Percentage")
plt.plot(bin_edges[1:],cdf)
plt.legend(('CDF'))
plt.show()
```

```python
counts, bin_edges = np.histogram(vr['petal_length'], bins = 10)

pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(('Pdf','Cdf'))
plt.title('PDF and CDF For iris virginica')
plt.xlabel("Petal length")
plt.ylabel("Percentage")
plt.show();
```
